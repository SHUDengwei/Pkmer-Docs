# 模块4｜学习仪表盘与周报生成

目标：统计练习数据、识别易错标签、导出 Markdown 周报。

```python
import pandas as pd, datetime as dt, os, json
path = '/mnt/data/qa_bank.csv'
assert os.path.exists(path), '未找到题库文件 /mnt/data/qa_bank.csv，请先完成模块3'
df = pd.read_csv(path)
df['result'] = pd.to_numeric(df.get('result', 0), errors='coerce').fillna(0)
acc = round(100 * df['result'].mean(), 1) if len(df) else 0.0
total = int(len(df))
done = int(df['result'].sum())
upcoming = df[df.get('next','') >= str(dt.date.today())]
# 易错标签统计
tag_mist = (df[df['result']==0]['tag'].value_counts().head(5)).to_dict() if 'tag' in df.columns else {}
report = f"""# 学习周报
- 完成题数: {done}/{total} (准确率 {acc}%)
- 易错标签Top: {tag_mist}
- 下周到期复习量: {len(upcoming)}
- 本周反思：
  - 我最容易错的知识点是……
  - 我准备通过……来改进
  - 下周我会在……时间段坚持练习
"""
        out_path = '/mnt/data/weekly_report.md'
        open(out_path, 'w', encoding='utf-8').write(report)
        print(report)
        print('\n已导出：', out_path)
```
