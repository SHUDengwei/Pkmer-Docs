## 步骤2：配置 Gemini API Key

### 如何获取API Key：

```python
import os
from google import genai
from google.genai import types

# 如果在 Colab 中运行
try:
    from google.colab import userdata
    api_key = userdata.get('GEMINI_API_KEY')
except ImportError:
    # 如果在本地运行，从环境变量读取
    api_key = os.getenv('GEMINI_API_KEY')

if not api_key:
    raise ValueError("❌ 请先配置 GEMINI_API_KEY")

client = genai.Client(api_key=api_key)
print("✅ Gemini API 配置成功！")
```

## 步骤3：测试 Gemini 连接


```python
# 测试：让AI鼓励你开始健康管理
response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents="请用一句话鼓励想要改善健康习惯的人！"
)

print("🤖 AI说：")
print(response.text)
```

## 步骤4：理解健康数据的四大维度

在AI健康管理中，我们主要关注四个方面：

### 🍎 1. 营养（Nutrition）
- **关键指标**：热量摄入、营养均衡度、三餐规律性
- **数据来源**：饮食日记、食物照片、热量记录
- **AI能做什么**：识别食物、计算营养、评估饮食健康度

### 😴 2. 睡眠（Sleep）
- **关键指标**：睡眠时长、入睡时间、睡眠质量评分
- **数据来源**：睡眠手环、主观记录、作息日志
- **AI能做什么**：分析睡眠模式、识别睡眠问题、提供改善建议

### 🏃 3. 运动（Exercise）
- **关键指标**：步数、运动时长、运动类型、心率
- **数据来源**：计步器、运动App、健身记录
- **AI能做什么**：追踪运动习惯、建议运动计划、评估运动强度

### 😌 4. 压力（Stress）
- **关键指标**：情绪评分、压力水平、工作强度
- **数据来源**：心情日记、压力自评、工作时长
- **AI能做什么**：识别压力模式、提供减压建议、预警健康风险

## 步骤5：第一次健康数据分析

让我们用AI分析一段简单的健康记录

```python
# 示例：一天的健康记录
health_diary = """
今天早上7点起床，感觉有点累。
早餐吃了煎饼果子和豆浆。
上午工作比较忙，一直坐着，几乎没起来活动。
中午吃了公司食堂的盖浇饭，有点油腻。
下午开会压力很大，喝了3杯咖啡。
晚上加班到9点，叫了外卖麻辣烫。
回家后刷手机到凌晨1点才睡。
"""

# 让AI分析这段记录并给出评价
prompt = f"""
请分析以下健康记录，从营养、睡眠、运动、压力四个维度给出评价和改善建议：

{health_diary}

请用清晰的结构输出：
1. 各维度评分（0-100分）
2. 主要问题
3. 最容易实施的3条改善建议
"""

response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=prompt
)
print("🏥 AI健康分析：")
print(response.text)
```
