# 模块2｜个性化学习画像与计划生成

目标：基于学习者画像，自动生成两周学习路线。

```python
import os, json, google.generativeai as genai
from google.colab import userdata
os.environ['GEMINI_API_KEY'] = userdata.get('GEMINI_API_KEY')
assert os.getenv('GEMINI_API_KEY'), '请先在模块0中设置 GEMINI_API_KEY'
genai.configure(api_key=os.getenv('GEMINI_API_KEY'))
model = genai.GenerativeModel('gemini-2.5-pro')
```

### 1) 编辑你的学习画像

```python
profile = {
    "科目": "英语三级",
    "目标": "两周内通过考试，口语≥85",
    "薄弱点": ["听力细节", "长难句"],
    "学习时段": ["工作日 20:30-21:30", "周末 9:30-11:00"],
    "偏好": ["情景对话", "短视频讲解"],
    "已有基础": "能进行日常交流"
}
print(json.dumps(profile, ensure_ascii=False, indent=2))
```

### 2) 生成两周学习计划（按天到任务）

```python
plan_prompt = f'''你是学习教练。根据画像生成两周学习路线，要求：
1) 按天输出任务（<=3项/天，每项≤30分钟）
2) 每天附一条复盘问题
3) 给出一份每周回顾清单
学习者画像：{json.dumps(profile, ensure_ascii=False)}'''
plan = model.generate_content(plan_prompt).text
print(plan)
```
